### Exercise - Requests
- Using the following url(http://www.yes24.com/Product/Goods/96971128)
- Find the following information
  - 책제목
  - 저자
  - 출판사, 출간일
  - 판매가, 할인율

```python
import requests
url = 'http://www.yes24.com/Product/Goods/96971128'
r = requests.get(url)
html = r.text
html
```

```python
soup = BeautifulSoup(html, 'lxml')
soup
```

```python
# 책제목
soup.find('div', attrs={'class':'gd_titArea'}).h2.text
```

```python
# 책저자
pubArea = soup.find('span', attrs = {'class':'gd_pubArea'})
pubArea.find('span', attrs = {'class':'gd_auth'}).text
```

```python
# 출판사
pubArea
```

```python
pubArea.find('span',attrs = {'class':'gd_pub'}).text
```

```python
# 출간일
pubArea.find('span', attrs = {'class':'gd_date'}).text
```

```python
# 판매가, 할인율
price_info = soup.find('div', attrs = {'class', 'gd_infoTb'})
price_info
```

```python
price_info.find_all('em', attrs = {'class':'yes_m'})
```

```python
price_info.find_all('em', attrs = {'class':'yes_m'})[0].text
```

```python
# 할인율
discount = price_info.find('tr', attrs = {'class':'accentRow'}).contents[3]
```

```python
discount.text[11:17]
```

```python
# 혹은
one_price = int(price_info.find_all('em',attrs = {'class':'yes_m'})[0].text[:-1].replace(',',''))
discount_price = int(price_info.find_all('em', attrs= {'class':'yes_m'})[1].text.replace(',',''))
```

```python
round((1-(discount_price/one_price))*100)
```
