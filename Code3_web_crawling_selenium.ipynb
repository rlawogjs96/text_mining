# Google app store review crawling

### 1. Chromedriver에 원하는 사이트 접속하기

```python
wanted_url = 'https://play.google.com/store/apps/details?id=com.disney.disneyplus&hl=ko&gl=US&showAllReviews=true'
driver.get(wanted_url)
time.sleep(2)
```

### 2. 화면 스크롤 내리기

```python
# Scroll 횟수 정하기
scroll_cnt = 5

for i in range(scroll_cnt):
    # '더보기' 버튼이 있는지 확인
    button_list = driver.find_elements(By.CLASS_NAME, 'RveJcd.snByac')
    
    if len(button_list) != 0:
        driver.find_element(By.CLASS_NAME, 'RveJvd.snByac').click()
        driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
    else:
        driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')

    time.sleep(2)
```

```python
review_box_list = driver.find_elements(By.CLASS_NAME, 'd15Mdf.bAhLNe')
print('리뷰 개수 :', len(review_box_list))
```

### 3. 리뷰 게시물 데이터

```python
f = open('google_app_review.txt', 'w', encoding = 'utf-8')

progress = 0

for review_box in review_box_list:
    progress += 1
    print(progress, end='\r')
    
    # 리뷰 별점 가져오기
    review_star = review_box.find_element(By.CLASS_NAME, 'pf5lIe')
    review_star_count = len(review_star.find_elements(By.CLASS_NAME, "vQHuPe.bUWb7c"))
    
    # 리뷰 작성일자 가져오기
    review_date = review_box.find_element(By.CLASS_NAME, 'p2TkOb').text
    
    # 리뷰 텍스트 가져오기
    try:
        review_box.find_element(By.CLASS_NAME, 'LkLjZd.ScJHi.OzU4dc').click()
        time.sleep(1)
        review_text = review_box.find_element(By.CLASS_NAME, 'UD7Dzf').text
        review_text = review_text.replace('\n',' ').strip()
    except:
        review_text = review_box.find_element(By.CLASS_NAME, 'UD7Dzf').text
        review_text = review_text.replace('\n',' ').strip()
    
    # 파일에 쓰기
    f.write(review_date + '\t'+ str(review_star_count)+ '\t'+review_text+'\n')
f.close()
```

```python
driver.close()
```

# Naver News crawling

```python
query_txt = input('검색할 키워드 :')
start_date = input('조회 시작 날짜 (예:20210101):')
end_date = input('조회 종료 날짜 9(예:20211231):')
```

```python
chrome_options = webdriver.ChromeOptions()

chrome_options.add_argument('--user-agent="Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/66.0.3359.139 Safari/537.36"')

# Windows 운영체제
#driver = webdriver.Chrome(executable_path = "chromedriver", chrome_options=chrome_options)

# MAC, Linux 운영체제
driver = webdriver.Chrome(executable_path = "/Users/gahyekim/Downloads/chromedriver", chrome_options=chrome_options)
```

### 1. 네이버 키워드 검색 후 뉴스 카테고리 선택

```python
driver.get('http://www.naver.com')
time.sleep(2)
```

```python
# 네이버 검색창에 '키워드' 입력
element = driver.find_element_by_id('query')
element.send_keys(query_txt)
element.submit()
```

```python
# '뉴스' 카테고리 선택
driver.find_element(By.LINK_TEXT, '뉴스').click()
```

### 2. 설정 날짜에 맞는 뉴스 찾기

```python
# '옵션' 열기
driver.find_element(By.CLASS_NAME, 'btn_option._search_option_open_btn').click()
```

```python
# '검색기간'- 직접입력
driver.find_element(By.CLASS_NAME, 'txt.txt_option._calendar_select_trigger').click()
```

```python
# 시작날짜 설정
driver.find_element(By.CLASS_NAME, "spnew_bf.ico_calendar._start_trigger").click()
date_list = driver.find_elements(By.CLASS_NAME, "lst_item._ul")
# 년
date_list[1].find_element(By.LINK_TEXT, str(start_date[0:4])).click()

# 월
if start_date[4:6].startswith("0"):
    date_list[2].find_element(By.LINK_TEXT, str(start_date[5:6])).click()
else:
    date_list[2].find_element(By.LINK_TEXT, str(start_date[4:6])).click()

# 일
if start_date[6:].startswith("0"):
    date_list[3].find_element(By.LINK_TEXT, str(start_date[7:])).click()
else:
    date_list[3].find_element(By.LINK_TEXT, str(start_date[6:])).click()
```

```python
# 종료날짜 설정
driver.find_element(By.CLASS_NAME, "spnew_bf.ico_calendar._end_trigger").click()
end_date_list = driver.find_elements(By.CLASS_NAME, "lst_item._ul")
# 년
end_date_list[1].find_element(By.LINK_TEXT, str(end_date[0:4])).click()

# 월
if end_date[4:6].startswith("0"):
    end_date_list[2].find_element(By.LINK_TEXT, str(end_date[5:6])).click()
else:
    end_date_list[2].find_element(By.LINK_TEXT, str(end_date[4:6])).click()

# 일
if end_date[6:].startswith("0"):
    end_date_list[3].find_element(By.LINK_TEXT, str(end_date[7:])).click()
else:
    end_date_list[3].find_element(By.LINK_TEXT, str(end_date[6:])).click()
```

```python
# '적용하기' 누르기
driver.find_element(By.CLASS_NAME, "btn_apply._apply_btn").click()
```

### 3. 네이버 뉴스 url 데이터 가져오기

```python
article_url = []

box_list = driver.find_elements(By.CLASS_NAME, 'info_group')
for an_new in box_list:
    naver_list = an_new.find_elements(By.LINK_TEXT, '네이버뉴스')
    if len(naver_list) != 0:
        url = an_new.find_element(By.LINK_TEXT, '네이버뉴스').get_attribute('href')
        article_url.append(url)
for i in range(10): # 검색 결과 10page까지 크롤링
    i += 1
    print(i, end='\r')
    btn_next = driver.find_elements(By.CLASS_NAME, 'btn_next')
    btn_next[-1].click()
    time.sleep(5)
    
    box_list = driver.find_elements(By.CLASS_NAME, 'info_group')
    for an_new in box_list:
        naver_list = an_new.find_elements(By.LINK_TEXT, '네이버뉴스')
        if len(naver_list) != 0:
            url = an_new.find_element(By.LINK_TEXT, '네이버뉴스').get_attribute('href')
            article_url.append(url)
```

### 4. 네이버 뉴스 본문 데이터 가져오기

```python
f = open('article_data.txt', 'w', encoding= 'utf-8')
progress = 0
non_url = []

for url in article_url:
    driver.get(url)
    progress+=1
    print(progress, end='\r')
    time.sleep(2)
    
    try:
        title = driver.find_element(By.CLASS_NAME, 'tts_head').text
        date_list = driver.find_elements(By.CLASS_NAME, 't11')
        date = date_list[0].text
        contents = driver.find_element(By.ID, 'articleBodyContents').text
        contents = contents.replace('\n', ' ')
        f.write(title+'\t'+date+'\t'+contents+'\n')
    except:
        non_url.append(url)
        continue
f.close()
```

```python
f = open('article_data.txt', 'r', encoding = 'utf-8')

for aline in f:
    print(aline.strip())
f.close()
```
